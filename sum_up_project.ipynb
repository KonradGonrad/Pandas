```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('adult.csv')
df.head()
```

```python
df.describe(include='all')
```

```python
df.info(verbose=True, show_counts=False)
```

---

```python
for col in df.select_dtypes(include=['object']).columns.tolist():
    print(f"{col}: {df[f'{col}'].unique()} | {len(df[f'{col}'].unique())}") # '?' is the issue

cols_del = df.columns[df.isin(values=['?', '', ' ']).any()] # Columns that has wrong values in there
print(f"Cols, where suspended values append: {cols_del.tolist()}")
```

```python
print(f"indexes of duplicates{df[df.duplicated(keep=False)].index}. Length: {len(df[df.duplicated(keep=False)].index)}")

age_duplicates = df[df.duplicated(keep=False)].age.unique().tolist()
df[df.duplicated(keep=False)].loc[df['age'] == age_duplicates[1]] # Example of duplicated rows
```

```python
b_nan = df.isna().sum().sum()
df.replace(to_replace=['?'],
           value=np.nan,
           inplace=True)
a_nan = df.isna().sum().sum()

print(f"Before nans: {b_nan}\nAfter nan: {a_nan}")
```

```python
df_size_before = len(df)
```

```python
b_dup = df.duplicated().any()
df.drop_duplicates(inplace=True)
a_dup = df.duplicated().any()

print(f"Before duplicates: {b_dup}\nAfter duplicates: {a_dup}")
```

```python
df.dropna(inplace=True)
a2_nan = df.isna().sum().sum()
print(f"After deletion: {a2_nan}")
```

```python
df_size_after = len(df)
print(f"Size before deletion: {df_size_before} -> Size after deletion: {df_size_after}. The percentage of deleted data: {((df_size_before - df_size_after)/df_size_before)*100:.2f} %")
```

---

```python
obj_cols = df.select_dtypes(include=['object']).columns
obj_cols
```

```python
country_counts = df['native.country'].value_counts() # liczba wystapien na narodowosc
country_counts
```

```python
rare_countries = country_counts[country_counts < 100].index
print(len(rare_countries))

df['native.country'] = df['native.country'].replace(to_replace=rare_countries, value='Other')
df['native.country'].unique()
```

```python
df['income'] = df['income'].map({'<=50K': 0, '>50K': 1})
X = df.drop('income', axis=1)
y = df['income']
```

```python
from sklearn.preprocessing import LabelEncoder

linear_model = False # Modele liniowe lub drzewiaste
swap_to_category = True # Zmiana typu danych na category
enc_cols = ['sex', 'workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'native.country']

if linear_model:
    X = pd.get_dummies(X, columns=enc_cols)
else:
    # X['sex'] = X['sex'].astype(dtype='category')
    # df['income'] = df['income'].astype(dtype='category')
    for col in enc_cols:
        X[col] = LabelEncoder().fit_transform(X[col])
        if swap_to_category:
            X[col] = X[col].astype(dtype='category')
    
X.head(5)
```

```python
int_cols = df.select_dtypes(include=['int64']).columns
for col in int_cols:
    df[col] = pd.to_numeric(df[col], downcast='integer')

df.dtypes
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

for int_col in int_cols:
    print(f"{int_col} | \n{df[int_col].agg(['min', 'max'])}\n")

int_col_scale = ['age', 'fnlwgt', 'capital.gain', 'capital.loss', 'education.num', 'hours.per.week']
X[int_col_scale] = scaler.fit_transform(X[int_col_scale])
X.head()
```

---

```python
print(f"1: {len(y.loc[y == 1])} |  0: {len(y.loc[y == 0])}") # inbalance
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_resampled, y_resampled = smote.fit_resample(X, y)

print(f"1: {len(y_resampled.loc[y_resampled == 1])} |  0: {len(y_resampled.loc[y_resampled == 0])}") # Balanced
```

---

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression

if linear_model:
    model = LogisticRegression(max_iter= 1000, random_state=42)
else:
    model = DecisionTreeClassifier(random_state=42)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_true=y_test, y_pred=y_pred)
print(f"Accuracy: {acc:.2f}")
```
